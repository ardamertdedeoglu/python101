```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
df = pd.read_csv("Housing.csv")
```

# SORU 1


```python
plt.scatter(df['area'], df['price'],alpha=0.5,c='blue')
plt.title("Alan ve Fiyatlar Arasndaki liki")
plt.xlabel("Blge")
plt.ylabel("Fiyat")
plt.gca().get_yaxis().get_major_formatter().set_useOffset(False)
plt.gca().get_yaxis().get_major_formatter().set_scientific(False)
plt.show()

plt.scatter(df['bedrooms'], df['price'],alpha=0.5,c='blue')
plt.title("Yatak Odas ve Fiyatlar Arasndaki liki")
plt.xlabel("Yatak Odas")
plt.ylabel("Fiyat")
plt.gca().get_yaxis().get_major_formatter().set_useOffset(False)
plt.gca().get_yaxis().get_major_formatter().set_scientific(False)
plt.show()

plt.scatter(df['stories'], df['price'],alpha=0.5,c='blue')
plt.title("Kat ve Fiyatlar Arasndaki liki")
plt.xlabel("Kat")
plt.ylabel("Fiyat")
plt.gca().get_yaxis().get_major_formatter().set_useOffset(False)
plt.gca().get_yaxis().get_major_formatter().set_scientific(False)
plt.show()
```

# SORU 2


```python
banyo_fiyatlar = df.groupby('bathrooms')['price'].mean()
plt.bar(banyo_fiyatlar.index, banyo_fiyatlar.values,color='skyblue')
plt.title("Banyo miktarna gre Ortalama Ev Fiyatlar")
plt.xlabel("Banyo Says")
plt.ylabel("Ortalama Fiyat")
plt.gca().get_yaxis().get_major_formatter().set_useOffset(False)
plt.gca().get_yaxis().get_major_formatter().set_scientific(False)
plt.show()

furnish_fiyatlar = df.groupby('furnishingstatus')['price'].mean()
plt.bar(furnish_fiyatlar.index, furnish_fiyatlar.values,color='skyblue')
plt.title("Mobilya Miktarna gre Ortalama Ev Fiyatlar")
plt.xlabel("Mobilyallk Oran")
plt.ylabel("Ortalama Fiyat")
plt.gca().get_yaxis().get_major_formatter().set_useOffset(False)
plt.gca().get_yaxis().get_major_formatter().set_scientific(False)
plt.show()
```

# SORU 3


```python
kategorik_sutunlar = df.select_dtypes(include=["object"]).columns
le= LabelEncoder()
for sutun in kategorik_sutunlar:
    df[sutun] = le.fit_transform(df[sutun])

korelasyon_matrisi = df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(korelasyon_matrisi,annot=True,cmap='coolwarm', fmt='.2f')
plt.title("Stunlar Aras Korelasyon Matrisi")
plt.show()
```
